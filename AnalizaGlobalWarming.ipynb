```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/temp/analiza_danych/Environment_Temperature_change_E_All_Data_NOFLAG.csv',encoding='latin-1')
```

```python
data
```

```python

poland=data.query('Area=="Poland"')
year_pl=poland.query('Months=="Meteorological year"')
year_pl=year_pl.loc[:, 'Y1961':'Y2019']
year_pl.reset_index(drop=True, inplace=True)
pl_show=np.array(year_pl)
colnames=list(year_pl.columns)
year_pl
```

```python
colnames2=[]
for year in colnames:
    colnames2.append(int(year.strip('Y')))
```

```python
plt.figure(figsize=(10,6))
plt.title("Change in the average temperature in Poland from 1961 to 2019 ",fontsize=16) 
plt.plot(colnames2,pl_show[0],color='darkorange')
plt.grid()
plt.show()
;
```

```python
world=data.query('Area=="OECD"')
year_world=world.query('Months=="Meteorological year"')
year_world=year_world.loc[:, 'Y1961':'Y2019']
year_world.reset_index(drop=True, inplace=True)
show=np.array(year_world)
colnames=list(year_world.columns)
year_world
```

```python
colnames2=[]
for year in colnames:
    colnames2.append(int(year.strip('Y')))
```

```python
plt.figure(figsize=(10,6))
plt.title("Change in the average temperature in world from 1961 to 2019 ",fontsize=16) 
plt.plot(colnames2,show[0],color='darkorange')
plt.grid()
plt.show()
;
```
